# Bronze medal: Integral of a Polynomial

a) Student A carefully examine his data and proposed a model: 

The clicks per hour with respect to time can be described by function 

## $y=7.98 \times 10^{-5} \cdot x^5 - 1.80 \times 10^{-2} \cdot x^4 + 1.48 \cdot x^3 - 52.5 \cdot x^2 + 680 \cdot x$

where x is the past time in hour and y is the clicks. Can you plot the data and this function to have a look?


```python
import numpy as np
import matplotlib.pyplot as plt

def plot_polynomial():
    def y(x):
        return (7.98e-5 * x**5 
                - 1.80e-2 * x**4 
                + 1.48 * x**3 
                - 52.5 * x**2 
                + 680 * x)
    
    x = np.linspace(0, 80,)  # Adjust range as needed
    y_values = y(x)
    
    plt.plot(x, y_values, label=r'$y = 7.98 \cdot 10^{-5} x^5 - 1.80 \cdot 10^{-2} x^4 + 1.48 x^3 - 52.5 x^2 + 680 x$')
    plt.xlabel('Past time (hours)')
    plt.ylabel('CLicks')
    plt.title('Polynomial Plot')
    plt.grid(True)
    plt.legend()
    plt.show()

# Call the function
plot_polynomial()
```

# Silver medal: Find the derivative of a function using the product rule and the rule of composite function

a) Student B carefully examined his data and checked the literature. After that, he use this function to fit his data:

## $y=144000\cdot e^{-x/8}\left( -x - 8\right)+1152000$

where x is the past time in hour and y is the total viewing time (in the unit of minus). 

Can you find the deriviative of this function?

```python
import sympy as sp

x = sp.Symbol('x')

u = 144000 * sp.exp(-x/8)
u_prime = -18000 * sp.exp(-x/8)  
v = -x - 8
v_prime = -1

y_prime = u_prime * v + u * v_prime

y_prime_simplified = sp.simplify(y_prime)

print("y'(x) =", y_prime_simplified)
```

# Gold medal: Find the derivative of a function using the quotient rule and conducting optimization

a) Student C found their work very interesting and got involved. In some other paper, Student C had read that the Sigmoid function was very useful for this type of data and he decided to create such a model. After a regression to get the constants, his function was::

## $f(x) = \left(\frac{1.65 \times 10^6} {1 + \exp{\left(-\frac{x}{8} + 1\right)}} \right)-5\times 10^5$

Can you help Student C find the function that demonstrates clicks per hour as a function of time?

```python

# Sigmoid-based clicks per hour (Student C)
def clicks_sigmoid(x):
    A = 1.65e6
    numerator = np.exp(-x/8 + 1)
    denominator = (1 + numerator)**2
    return (A / 144) * numerator / denominator  # 144 = 18*8

# Function to plot Student A data and Student C model
def plot_sigmoid_comparison(csv_file='data_student_A.csv'):
    # Load Student A data
    data = np.genfromtxt(csv_file, delimiter=',', skip_header=1, dtype=float)
    time_A = data[:, 0]
    clicks_A = data[:, 1]

    x_vals = np.linspace(min(time_A), max(time_A), 500)
    clicks_C = clicks_sigmoid(x_vals)

    plt.scatter(time_A, clicks_A, color='red', label="Student A data", alpha=0.7)
    plt.plot(x_vals, clicks_C, color='green', label="Sigmoid model (Student C)")
    plt.xlabel("Time (hours)")
    plt.ylabel("Clicks per hour")
    plt.title("Student A Data vs Sigmoid Model (Student C)")
    plt.grid(True)
    plt.legend()
    plt.show()

# Call the function
plot_sigmoid_comparison()
```

c) Use the best model to calculate: at which hour does clicks per hour reach its maximum value? (rounded to two decimal places)

```python
from scipy.optimize import minimize_scalar

def find_max_clicks(model):
    # maximize f(x) by minimizing -f(x)
    res = minimize_scalar(lambda x: -model(x), bounds=(0, 100), method='bounded')
    return res.x, model(res.x)

# Find maximum
t_max, clicks_max = find_max_clicks(BestModel)

print("The maximum clicks per hour occur at {:.2f} hours with {:.2f} clicks/hour.".format(t_max, clicks_max))
```

d) This YouTuber happened to get a double-coupon. This special double-coupon can doubles his profits from all viewings for 5 hours and he can choose the exact time to start. Based on the model we found, can you suggest which hour he should start using this coupon? The time is counted as 0 when his video is published. (rounded to two decimal places)

```python
import numpy as np
from scipy.integrate import quad
from scipy.optimize import minimize_scalar

def DerivativeExp(x):
    return 1000 * x * np.exp(-x/8)

BestModel = DerivativeExp

def find_best_coupon_time(model, duration=5, max_time=80):
    # Function to minimize (negative total clicks over the window)
    def negative_total_clicks(t):
        total, _ = quad(model, t, min(t + duration, max_time))
        return -total

    # Optimize start time t between 0 and max_time-duration
    res = minimize_scalar(negative_total_clicks, bounds=(0, max_time - duration), method='bounded')
    return res.x

# Find best 5-hour period to activate the double coupon
best_start = find_best_coupon_time(BestModel)
print("The best time to start the 5-hour double coupon is at hour {:.2f}".format(best_start))
```
